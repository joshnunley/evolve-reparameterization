```julia
using Graphs
using MetaGraphs

using Graphs: SimpleDiGraph
```

```julia
struct LocalDAG
    g::SimpleDiGraph
end

function LocalDAG(
        input_size,
        output_size,
        num_hidden,
        window_size
    )
    num_nodes = (num_hidden + 1)*input_size + output_size
    g = SimpleDiGraph(num_nodes)
    
    
    return LocalDAG(g)
end
```

```julia
struct BinaryFunction
    weights::Vector{Ref{Float64}}
    bin_to_dec::Function
    threshold::Function
    function BinaryFunction(weights)
        function bin_to_dec(x)
            return sum([x[end-i]*2^i for i in 0:(length(x)-1)])
        end

        function threshold(x)
            return (x < 0) ? 0 : 1
        end
        new(weights, bin_to_dec, threshold)
    end
end

function (m::BinaryFunction)(x)
    return m.threshold(m.weights[m.bin_to_dec(x)+1][]), m.weights[m.bin_to_dec(x)+1]
end
```

```julia
struct LocalBinaryNN
    eval::Matrix{Bool}
    hidden_funcs::Matrix{BinaryFunction}
    output_func::BinaryFunction
    input_size::Int64
    num_hidden::Int64
    window_size::Int64
    window_center::Int64
    output_sample::Vector{Int64}
    utilized_weights::Vector{Ref{Float64}}
end

function LocalBinaryNN(
        input_size,
        output_size, # currently unused, assumed single bit output
        num_hidden,
        window_size,
        window_center,
        output_sample_size
    )
    matrix_size = (num_hidden + 1, input_size)
    
    eval = zeros(Bool, matrix_size)
    
    # Broadcast piping ensures the random vectors are different
    hidden_funcs = [BinaryFunction(randn(2^window_size)) for i in 1:num_hidden, j in 1:input_size] 
    output_func = BinaryFunction(randn(2^output_sample_size))
    output_sample = rand(1:input_size, output_sample_size)
    
    utilized_weights = zeros(num_hidden*input_size + 1)
    
    return LocalBinaryNN(
                    eval, 
                    hidden_funcs, 
                    output_func, 
                    input_size, 
                    num_hidden, 
                    window_size, 
                    window_center, 
                    output_sample,
                    utilized_weights
            )
end

function (m::LocalBinaryNN)(x)
    m.eval[1, :] = x

    function index_wrap(ind, last)
        if (ind >= 1) && (ind <= last)
            return ind
        elseif (ind <= 0)
            return last + ind
        else
            return ind - last
        end
    end
    
    utilized_index = 1
    for i in 1:m.num_hidden
        for j in 1:m.input_size
            window_indices = (1:m.window_size) .- m.window_center
            window = [m.eval[i, index_wrap(j + k, end)] for k in window_indices]
            m.eval[i + 1, j], m.utilized_weights[utilized_index] = m.hidden_funcs[i, j](window)
            
            utilized_index += 1
        end
    end
    
    output, m.utilized_weights[end] = m.output_func(m.eval[end, m.output_sample])
    
    return output
    
end
```

```julia
input_size = 784
output_size = 1
num_hidden = 10
window_size = 3
window_center = 2
output_sample_size = 12

nn = LocalBinaryNN(
        input_size,
        output_size,
        num_hidden,
        window_size,
        window_center,
        output_sample_size
    )
```

```julia
function update!(utilized_weights::Vector{Ref{Float64}}, epsilon::Vector{Float64})
    for i in 1:length(utilized_weights)
        weight = utilized_weights[i]
        eps = epsilon[i]
        
        weight_sign = sign(weight[])
        weight[] = weight_sign*(abs(weight[]) + eps)
    end
end

function update!(utilized_weights::Vector{Ref{Float64}}, epsilon_mean::Float64, epsilon_variance::Float64)
    epsilon = epsilon_variance.*randn(length(utilized_weights)) .+ epsilon_mean
    update!(utilized_weights, epsilon)

end
```

```julia
using MLDatasets
using Flux: onehot
using Plots
using StatsBase: sample, mean
```

```julia
# load full training set
train_x, train_y = MNIST.traindata();

function binary_label(num)
    if num == 0
        return 1
    else
        return 0
    end
end

train_x_vec = [Int.(round.(vec(train_x[:, :, i]))) for i = 1:60000];
#train_y_hot = [onehot(train_y[i], 0:9) for i = 1:60000];
train_y_bin = [binary_label(train_y[i]) for i = 1:60000];

# load full test set
test_x,  test_y  = MNIST.testdata();

test_x_vec = [Int.(round.(vec(test_x[:, :, i]))) for i = 1:10000];
#test_y_hot = [onehot(test_y[i], 0:9) for i = 1:10000];
test_y_bin = [binary_label(test_y[i]) for i = 1:10000];

traindata = [(train_x_vec[i], train_y_bin[i]) for i = 1:60000];
testdata = [(test_x_vec[i], test_y_bin[i]) for i = 1:10000];
```

```julia
#sample_size = 1000
#train_sample = sample(traindata, sample_size, replace=false);
#test_sample = sample(testdata, sample_size, replace=false);

train_sample = [traindata[train_y_bin .== 1][1:500]; traindata[train_y_bin .== 0][1:500]];
test_sample = [testdata[test_y_bin .== 1][1:500]; testdata[test_y_bin .== 0][1:500]];
```

```julia
for x in train_sample
    guess = nn(x[1]);
    
    if guess == x[2]
        update!(nn.utilized_weights, 1.0, 1.0)
    else
        update!(nn.utilized_weights, -1.0, 0.1)
    end
end

correct = zeros(Bool, sample_size)
for i in 1:sample_size
    x = train_sample[i]
    guess = nn(x[1]);
    
    if guess == x[2]
        correct[i] = 1
        #println(guess, train_y[i])
    end
end
println(sum(correct)/sample_size)
println(1-sum([x[2] for x in train_sample])/sample_size)
```

```julia
num_weights = length(nn.utilized_weights)
weights_change = zeros(num_weights)
for x in train_sample
    guess = nn(x[1]);
    
    if guess == x[2]
        weights_change += 0.25.*randn(num_weights) .+ 1.0
    else
        weights_change += 1.0.*randn(num_weights) .+ -1
    end
end

update!(nn.utilized_weights, weights_change)

correct = zeros(sample_size)
for i in 1:sample_size
    x = train_sample[i]
    guess = nn(x[1]);
    
    if guess == x[2]
        correct[i] = 1
    end
end
println(sum(correct)/sample_size)
```
